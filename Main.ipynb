```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score, pairwise_distances
from collections import Counter
from itertools import chain
from sklearn.preprocessing import LabelEncoder
```

```python
def load_opcode_data(folder_path):
    program_opcodes = {}
    for file in os.listdir(folder_path):
        if file.endswith(".txt"):  # Assembly files are in .txt format
            with open(os.path.join(folder_path, file), "r") as f:
                opcodes = f.read().split()  # Splitting based on whitespace
                program_opcodes[file] = opcodes
    return program_opcodes

# Load opcode data from the folder
folder_path = "./Example Programs for Kernel Methods HMK 2" 
program_opcodes = load_opcode_data(folder_path)
print(program_opcodes)
```

```python
def build_dictionary(opcode_data):
    all_opcodes = list(chain.from_iterable(opcode_data.values()))  # Flatten list
    dictionary = sorted(set(all_opcodes))  # Unique sorted opcodes
    return dictionary

def remove_stop_words(opcode_data, dictionary, threshold=0.5):
    total_docs = len(opcode_data)
    opcode_counts = Counter(chain.from_iterable(set(opcodes) for opcodes in opcode_data.values()))
    
    # Identify stop words
    stop_words = {opcode for opcode, count in opcode_counts.items() if count / total_docs > threshold}
    filtered_dictionary = [opcode for opcode in dictionary if opcode not in stop_words]
    
    return filtered_dictionary
    
def encode_programs(opcode_data, dictionary):
    feature_matrix = np.zeros((len(opcode_data), len(dictionary)))
    for i, (file, opcodes) in enumerate(opcode_data.items()):
        for opcode in opcodes:
            if opcode in dictionary:
                feature_matrix[i, dictionary.index(opcode)] += 1
    return feature_matrix

def compute_kernels(X):
    dot_kernel = np.dot(X, X.T)  # Dot product kernel
    poly_kernel = (dot_kernel + 1) ** 2  # Polynomial kernel (d=2)
    
    return dot_kernel, poly_kernel
```

```python
# Create dictionary D0
D0 = build_dictionary(program_opcodes)
print(f"Size of Dictionary D0: {len(D0)}")

# Create dictionary D0 after removing stop words
D0_filtered = remove_stop_words(program_opcodes, D0)
print(f"Size of Filtered Dictionary D0: {len(D0_filtered)}")

#Representing each document as a vector
# Encode programs into feature matrix
X_D0 = encode_programs(program_opcodes, D0_filtered)
print(f"Feature Matrix Shape: {X_D0.shape}")

print(X_D0)
```

```python
# Compute kernels for D0
dot_kernel_D0, poly_kernel_D0 = compute_kernels(X_D0)
print("Dot Product Kernel: ", dot_kernel_D0.shape)
print(dot_kernel_D0)
print("Polynomial Kernel:", poly_kernel_D0.shape)
print(poly_kernel_D0)
```

```python
#PART-1
# Diagonal mean (self-similarity)
dot_kernel_D0_diag_mean = np.mean(np.diag(dot_kernel_D0))

# Mean of entire matrix (overall similarity)
dot_kernel_D0_total_mean = np.mean(dot_kernel_D0)

# Visualizing as heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(dot_kernel_D0, annot=False, cmap="coolwarm")
plt.title("Dot Kernel D0 Matrix Heatmap")
plt.show()

print(f"Mean of diagonal elements (self-similarity): {dot_kernel_D0_diag_mean:.2f}")
print(f"Mean of entire kernel matrix: {dot_kernel_D0_total_mean:.2f}")
```

```python
# Diagonal mean (self-similarity)
poly_kernel_D0_diag_mean = np.mean(np.diag(poly_kernel_D0))

# Mean of entire matrix (overall similarity)
poly_kernel_D0_total_mean = np.mean(poly_kernel_D0)

# Visualizing as heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(poly_kernel_D0, annot=False, cmap="coolwarm")
plt.title("Poly Kernel D0 Matrix Heatmap")
plt.show()

print(f"Mean of diagonal elements (self-similarity): {poly_kernel_D0_diag_mean:.2f}")
print(f"Mean of entire kernel matrix: {poly_kernel_D0_total_mean:.2f}")
```

```python
def stem_opcodes(opcode_list):
    opcode_mapping = {
    "push": ["push", "pusha", "pushb"],
    "pop": ["pop", "popq"],
    "add": ["addf", "addi", "add"],
    "sub": ["subf", "subi", "sub"],
    "mult": ["multi", "multf", "mul"],
    "div": ["divf", "divi", "div"],
    "cmp": ["cmpf", "cmpi", "cmp"],
    "jmp": ["jmp", "jz", "jnz", "je", "jne", "jbe", "jb", "ja", "jge", "jl"],
    "mov": ["mov", "movi"],
    "load": ["load", "ld"],
    "store": ["store", "st"],
    "xor": ["xor", "xori"],
    "and": ["and", "andi"],
    "or": ["or", "ori"],
    "test": ["test", "tst"],
    "call": ["call", "callq"],
    "lea": ["lea", "leaq"],
    "shr": ["shr", "shrq"],
    "shl": ["shl", "shlq"],
    "retn": ["retn", "ret"],
    "stos": ["stosb", "stosd", "stosq"],
    "movs": ["movsb", "movsd", "movsq"],
    "cmpxchg": ["cmpxchg", "cmpxchg8b"],
    "dec": ["dec", "deci"],
    "inc": ["inc", "inci"],
    "sbb": ["sbb", "sbbi"],
    "neg": ["neg", "negq"],
    "not": ["not", "notq"],
    "rep": ["rep", "repe", "repne"],
    "scas": ["scas", "scasb"],
    "stos": ["stos", "stosb", "stosd"],
    "movzx": ["movzx", "movzxq"],
    "movsx": ["movsx", "movsxq"],
    "shld": ["shld", "shldq"],
    "shrd": ["shrd", "shrdq"],
    "bt": ["bt", "btr", "bts", "btc"],
    "set": ["seta", "setae", "setb", "setbe", "sete", "setg", "setge", "setl", "setle", "setne", "setnp", "setns", "setnz", "seto", "setp", "setpe", "setpo", "sets", "setz"],
    "cmov": ["cmova", "cmovae", "cmovb", "cmovbe", "cmove", "cmovg", "cmovge", "cmovl", "cmovle", "cmovne", "cmovnp", "cmovns", "cmovnz", "cmovo", "cmovp", "cmovpe", "cmovpo", "cmovs", "cmovz"],
    "imul": ["imul", "imulq"],
    "idiv": ["idiv", "idivq"]
}
    stemmed_opcodes = []
    for opcode in opcode_list:
        found = False
        for root, variations in opcode_mapping.items():
            if opcode in variations:
                stemmed_opcodes.append(root)
                found = True
                break
        if not found:
            stemmed_opcodes.append(opcode)  # Keep opcode as is if no match
    return stemmed_opcodes

def apply_stemming_to_programs(opcode_data):
    stemmed_opcode_data = {file: stem_opcodes(opcodes) for file, opcodes in opcode_data.items()}
    return stemmed_opcode_data
```

```python
opcode_generalization = {
    "arith": ["add", "addi", "addf", "sub", "subi", "subf", "mul", "multi", "multf", "div", "divf", "divi", "imul", "idiv", "neg", "inc", "dec", "sbb"],
    "go": ["jmp", "jz", "jnz", "je", "jne", "jbe", "jb", "ja", "jge", "jl", "call", "callq", "ret", "retn", "loop"],
    "logic": ["xor", "xori", "and", "andi", "or", "ori", "not", "notq", "test", "tst"],
    "data": ["mov", "movi", "movzx", "movsx", "lea", "leaq", "load", "ld", "store", "st"],
    "stack": ["push", "pusha", "pushb", "pop", "popq"],
    "bitwise": ["shl", "shlq", "shr", "shrq", "shld", "shrd", "bt", "btr", "bts", "btc"],
    "compare": ["cmp", "cmpf", "cmpi", "cmpxchg", "cmpxchg8b"],
    "string": ["stos", "stosb", "stosd", "stosq", "scas", "scasb", "movs", "movsb", "movsd", "movsq", "rep", "repne", "repe", "cmpsb"],
    "set": ["seta", "setae", "setb", "setbe", "sete", "setg", "setge", "setl", "setle", "setne", "setnp", "setns", "setnz", "seto", "setp", "setpe", "setpo", "sets", "setz"],
    "cmov": ["cmova", "cmovae", "cmovb", "cmovbe", "cmove", "cmovg", "cmovge", "cmovl", "cmovle", "cmovne", "cmovnp", "cmovns", "cmovnz", "cmovo", "cmovp", "cmovpe", "cmovpo", "cmovs", "cmovz"],
}
def generalize_opcodes(opcode_list):
    generalized_opcodes = []
    for opcode in opcode_list:
        found = False
        for category, op_list in opcode_generalization.items():
            if opcode in op_list:
                generalized_opcodes.append(category)
                found = True
                break
        if not found:
            generalized_opcodes.append(opcode)  # Keep opcode if no mapping exists
    return generalized_opcodes

def apply_generalization_to_programs(opcode_data):
    generalized_opcode_data = {file: generalize_opcodes(opcodes) for file, opcodes in opcode_data.items()}
    return generalized_opcode_data
```

```python
#PART-4
def convert_kernel_to_distance(kernel_matrix):
    # Convert similarity to distance
    distance_matrix = np.max(kernel_matrix) - kernel_matrix
    #distance_matrix = np.abs(distance_matrix)  # Ensure all values are positive
    np.fill_diagonal(distance_matrix, 0)
    return distance_matrix

def evaluate_clustering(kernel_matrix, cluster_labels):
    return silhouette_score(kernel_matrix, cluster_labels, metric="precomputed")

def apply_spectral_clustering(kernel_matrix, n_clusters):
    clustering = SpectralClustering(n_clusters=n_clusters, affinity='precomputed', random_state=42)
    cluster_labels = clustering.fit_predict(kernel_matrix)
    return cluster_labels

```

```python
# Define cluster numbers to test
cluster_nums = [2, 3, 4]

# Store results
clustering_results = {}

for dataset, kernel_matrix in zip(["D0", "D1", "D2"], [dot_kernel_D0, dot_kernel_D1, dot_kernel_D2]):
    clustering_results[dataset] = {}
    
    # Convert kernel to distance
    distance_matrix = convert_kernel_to_distance(kernel_matrix)

    for k in cluster_nums:
        print(f"\nProcessing {dataset} with {k} clusters")
        print("Original Kernel Shape:", kernel_matrix.shape)
        
        # Apply Spectral Clustering
        clusters = apply_spectral_clustering(kernel_matrix, n_clusters=k)

        # Evaluate Clustering
        silhouette = evaluate_clustering(distance_matrix, clusters)
        clustering_results[dataset][k] = silhouette

        print(f"Silhouette Score for {dataset} with {k} clusters: {silhouette:.4f}")
```

```python
# Convert results to DataFrame for visualization
df_results = pd.DataFrame(clustering_results)

# Plot results
plt.figure(figsize=(8, 5))
df_results.plot(kind='bar', figsize=(10, 6))
plt.title("Silhouette Scores for Spectral Clustering on D0, D1, and D2")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.xticks(rotation=0)
plt.legend(title="Dataset")
plt.grid()
plt.show()
```
